# Paso 1: Registro de vehículos

Los usuarios de los vehículos se dirigen al TA (Autoridad de Confianza) y proporcionan información personal, como detalles de la tarjeta Aadhar, número de teléfono, número de licencia de conducir, número de vehículo, etc. Esta información se almacena en la base de datos maestra del TA.
Utilizando esta información, el TA genera claves únicas necesarias para cada vehículo de usuario a través del proceso de generación de claves, incluyendo la generación de la identidad de usuario original (OID), la identidad de usuario ficticia (DID) y un número aleatorio almacenado en la base de datos local. El mapeo de identidades originales a identidades ficticias se realiza únicamente en el TA.
Después de eso, la clave de autenticación, que consiste en una identidad ficticia y un número aleatorio, se almacena en el dispositivo a prueba de manipulaciones (TPD) del vehículo. La identidad ficticia protege la privacidad del vehículo.
# Paso 2: Generación de claves

Los RSUs (Unidades de Carretera Inteligente) generan sus propios libros contables privados que almacenan los detalles recibidos de los RSUs vecinos, como el seudónimo del vehículo, su mensaje, estado de autenticación y marca de tiempo.
Los RSUs generan parámetros de autenticación a través del emparejamiento bilineal para curvas elípticas. Esto implica que el vehículo selecciona un número aleatorio como clave privada a corto plazo y calcula una clave pública a corto plazo correspondiente en la curva elíptica.
El RSU verifica si el vehículo está autenticado o no. Si el vehículo no está autenticado, el RSU solicita los parámetros de autenticación; de lo contrario, se pasa el estado de "autenticación exitosa".
# Paso 3: Generación de firma

Para preservar la integridad del mensaje, el vehículo genera una firma anónima de corta duración utilizando el algoritmo HMAC-SHA256 y la clave privada. El vehículo calcula el hash del mensaje utilizando el algoritmo HMAC-SHA256 y lo asigna a un punto en una curva elíptica. La firma del mensaje se calcula como S = x * H(M), donde S es la firma y x es la clave privada.
# Paso 4: Priorización, verificación y reducción de autenticaciones

El vehículo envía el mensaje, la firma y la clave pública al RSU en forma de mensaje = (M || S || P). El mensaje se prioriza según el tipo de mensaje M.
El RSU autentica al vehículo si e(G, S) = e(P, H(M)) es verdadero; de lo contrario, se considera un vehículo malintencionado y el RSU informa al TA. También se verifica la integridad del mensaje utilizando el algoritmo HMAC-SHA256.
# Paso 5: Seguimiento

Los RSUs almacenan mensajes útiles en la blockchain utilizando el algoritmo de consenso Proof of Authority (PoA) en forma de transacciones. En caso de disputa, el TA rastrea la identidad real mediante la búsqueda en su base de datos local y en la base de datos de la blockchain. Después de rastrear la identidad real, el TA revoca la privacidad del vehículo o RSU malintencionado para evitar daños adicionales.
